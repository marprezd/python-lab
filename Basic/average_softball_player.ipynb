```python
# Program: average_softball_player.ipynb
# Written by Mario Pérez
# Date: 02-07-2022
#
# Declaring global variables.
ask_jersey_number = "Enter player jersey number"
ask_first_name = "Enter player's first name"
ask_last_name = "Enter player's last name"
ask_number_bases_earned = "Enter number of bases earned"
ask_number_of_at_bats = "Enter number of at-bats"
ask_next_steep = "Enter '1' for a new jersey number, or '0' for exit!"
ask_player_hits = "Enter the player’s hits"
ask_walks = "Enter number of base on balls(walk)"
ask_sacrifice_flies = "Enter number of sacrifice flies"
heading_line = "-" * 10
heading_title = " Average record by softball's player "
softball_team = "Stengel County softball team"


def main():
    """This is the main function that require a few data of the performance
    for each player within a Softball team to calculate different stadistic
    average, as slugging, player’s on-base, and a gross production average
    (GPA).
    """
    global jersey_number
    start_app = True  # Flag to start the program.

    while start_app:
        # Get the first jersey number.
        jersey_number = int(input(ask_jersey_number + " >> "))

        # Call function
        get_details_player()

        # We ask if user want to add another jersey number. If yes,
        # the start_app statement runs as True and the while loop starts again,
        # otherwise, the start_app statement runs as False and does not
        # start the loop by ending the program.
        next_step = int(input(ask_next_steep + " >> "))
        if next_step != 0:
            start_app = True
        else:
            start_app = False
    # End of program.
    end_of_job()


def get_details_player():
    """ This function receive all details of each player on
    the sofball team, and display the output.
    """
    first_name = input(ask_first_name + " >> ").capitalize()
    last_name = input(ask_last_name + " >> ").capitalize()
    number_bases_earned = int(input(ask_number_bases_earned + " >> "))
    number_of_at_bats = int(input(ask_number_of_at_bats + " >> "))
    player_hits = int(input(ask_player_hits + " >> "))
    walks = int(input(ask_walks + " >> "))
    sacrificie_flies = int(input(ask_sacrifice_flies + " >> "))

    # Assign data value to output variable
    print()
    output = heading_line + heading_title + heading_line + '\n'
    output += f"{softball_team : >20}\n"
    output += f"Player's name:  {last_name}, {first_name}\n"
    output += f"Jersey number:{jersey_number: >11}\n"
    output += '---------------------------------------------------------\n'
    output += f"Number of bases earned:{number_bases_earned : >2}\n"
    output += f"Number of at-bats:{number_of_at_bats : >7}\n"
    output += f"Player's hits:{player_hits : >11}\n"
    output += f'Base on balls (walk):{walks : >4}\n'
    output += f'Sacrificie flies:{sacrificie_flies : >8}\n'
    # Assign the returned value of the functions call to the output
    # variable.
    output += calculate_slugging_average(number_bases_earned,
                                         number_of_at_bats)
    output += calculate_player_on_base(player_hits, walks,
                                       number_of_at_bats,
                                       sacrificie_flies)
    output += calculate_gross_production_average()
    output += '---------------------------------------------------------'

    # Display result
    print(output)


def calculate_slugging_average(base_earned, of_at_bats):
    """ This function compute slugging percentages for his players. Slugging
    average is the total bases earned with base hits (base_earned) divided by
    the player’s number of at-bats (of_at_bats).
    """
    global slugging_average
    slugging_average = base_earned / of_at_bats
    return f"Slugging average:{slugging_average:>15.1%}\n"


def calculate_player_on_base(hits, walks, of_at_bats, sacrificie):
    """This function calculate a player’s on-base percentage. An on-base per-
    centage is calculated by adding a player’s hits and walks, and then
    dividing by the sum of at-bats, walks, and sacrifice flies.
    """
    global on_base_average
    on_base_average = (hits + walks) / (walks + of_at_bats + sacrificie)
    return f"On-base Average:{on_base_average:>16.1%}\n"


def calculate_gross_production_average():
    """This function computes a gross production average (GPA) for each pla-
    yer. A GPA is calculated by multiplying a player’s on-base percentage by
    1.8, then adding the player’s slugging percentage, and then dividing
    by four"""
    gpa = ((on_base_average * 1.8) + slugging_average) / 4
    return f"Gross Production Average:{gpa:>7.1%}\n"


def end_of_job():
    """This function marks the end of the program.
    """
    end_of_program = 'Thanks for use the program.'
    print(end_of_program)


# Call the main function.
main()
```
